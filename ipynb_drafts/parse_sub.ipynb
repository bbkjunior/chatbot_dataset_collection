```python
#Create lemmatizer and stopwords list
#mystem = Mystem() 
russian_stopwords = stopwords.words("russian")

#Preprocess function
def preprocess_text_default(text):
    sentences = nltk.sent_tokenize(text)
    clean_sentence = []
    #print(sentences)
    for sentence in sentences:
        words = sentence.split()
        clean_text= ''
        for word in words:
            clean_word = ''
            for char in word:
                if char != " " and char not in full_punctuation:
                    clean_word += char.lower()
            clean_text += clean_word + ' '
        clean_text = re.sub(' +', ' ', clean_text)
        clean_text = clean_text.rstrip()
        clean_text = clean_text.lstrip()
        clean_sentence.append(clean_text)
    return clean_sentence
```

```python
def preprocess_text(line):
    clean_text= ''
    for word in line.split():
        clean_word = ''
        for char in word:
            if char != " " and char not in full_punctuation:
                clean_word += char.lower()
        clean_text += clean_word + ' '
    clean_text = re.sub(' +', ' ', clean_text)
    clean_text = clean_text.rstrip()
    clean_text = clean_text.lstrip()
    clean_sentence.append(clean_text)
    return clean_sentence
```

```python
def check_metastring(line):
    for char in line:
        if char != " " and char not in full_punctuation and char.isalpha() :
            return line
    return ''
```

```python
#Create lemmatizer and stopwords list
mystem = Mystem() 
russian_stopwords = stopwords.words("russian")

#Preprocess function
def preprocess_text(text):
    tokens = mystem.lemmatize(text.lower())
    tokens = [token for token in tokens if token not in russian_stopwords\
              and token != " " \
              and token.strip() not in punctuation]
    
    text = " ".join(tokens)
    
    return text
```

```python
food_word = ['кушать','покушать','еда','пища','блюдо','суп','пир','питье',
             'выпивка' ,'закуска','жратва','угощение','лакомство','ужин','завтрак','напиток',
            'щи','каша','борщ','похлебка','свинина','говядина','пирог','квашеный','котлета','мясо','баранина','курятина',
            'овощ', 'фрукт']

def detect_food(line):
    words = line.split()
    for word in words:
        for food in  food_word:
            try:
                sim = fasttext.similarity(word, food)
                if(sim > 0.8):
                    print(line, ":",word,  food, sim)
                    food_stc = [line, word, food]
                    return True, food_stc
            except:
                pass
                
    return False, None
        
```
